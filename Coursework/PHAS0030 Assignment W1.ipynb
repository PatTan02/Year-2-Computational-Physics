**Introduction**

The following Sections will be trying to plot the electric potential and field generated by charges placed in 3 different configuration (monopole/one, dipole/two of opposite signs, quadrupole/four with two positive, two negative) using the principle of superposition that the resultant field is equivalent to the vector sum of the electric field. The electric potential of a charge is given by $$V= \frac{q}{|\mathbf{r}|} $$ and the the electric field is given by $$
\mathbf{E}(\mathbf{r})=q \frac{\mathbf{r}}{|\mathbf{r} |^3} 
$$
where $$| \mathbf{r}| = \sqrt{(\mathbf{x} - \mathbf{x}_0)^2 +(\mathbf{y} - \mathbf{y}_0)^2 }$$ and $$(\mathbf{x}_0,\mathbf{y}_0)$$ is the position of the charge particle

First import matplot and numpy library 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### Electric Potential Section
Start by defining the potential function,then the position of the charge, and the range of value x and y. Using np.meshgrid to generate 2D arrays to be used as input for the potential function to map out the whole potential across all possible value of position across the possible range of x and y. Finally, Plot the monopole, dipole and quadrupole graph

```python
def potential(x0,y0,x,y,q):
    """This function calculates the electric potential for a set of given input parameters 
    
    Inputs:
    q  Value of charge
    x0, y0  Location of charge (floats)
    x, y    Position(s) for potential to be evaluated (2D arrays)
    """
    r = np.sqrt((x-x0)**2 + (y-y0)**2)
    potential = q/r
    return potential 
```

```python
"""Case 1: Monopole (one charge particle)"""

# Define variables for charge(s) and positions
q1=-2
x1 = 2
y1 = 2
# Create grid of x and y positions for plotting
x=np.arange(0,10,1)
y=np.arange(0,10,1)

#generate 2D coordinate arrays for later vectorised evaluation of 2D potential, field
xx,yy=np.meshgrid(x,y)

# Call function and calculate the potential field (2D array)
pot1 = potential(x1,y1,xx,yy,q1) 
ax=plt.imshow(pot1) 
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title(" Monopole Potential")

plt.figure(figsize=(30, 30))
```

```python
"Case 2: Dipole (add a second charge with opposite sign)"""
#Calculating the potential of charged particle 2
q2 = 2
x2 = 8
y2 = 2
pot2 = potential (x2,y2,xx,yy,q2) 

# Calculate the total potential by both charged particle and plot the potential field
dipole_pot= pot1+pot2  
plt.imshow(dipole_pot)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dipole Potential')
plt.colorbar()
```

```python
"""Case 3: Quadrupole (add two more charges arranged in a square, 
        with like charges diagonally oppposite each other )"""
#Calculation of potential for charge 3 and 4
q3 = - 2
x3 = 8
y3 = 8
pot3 = potential (x3,y3,xx,yy,q3)


q4 = 2
x4 = 2
y4 = 8
pot4 = potential (x4,y4,xx,yy,q4)

# Calculation of the final potential for all charge and Plot the potential
quadrupole_pot= pot1+pot2+pot3+pot4
plt.imshow(quadrupole_pot)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadruple Potential')
plt.colorbar()
```

### Section on field
Ploting the field for Monopole, Dipole and Quadruple configuration of charges 

```python
def electric_field(x0,y0,x,y,q):
    '''Computing each individual component of the electric generated by the charged particle
    
    inputs:
    q - value of charge
    x0, y0  Location of charge (floats)
    x, y    Position(s) for potential to be evaluated
    '''
    
    r = abs(np.sqrt((x-x0)**2 + (y-y0)**2))
    Ex = q*(x-x0)/(r**3)                    # x component of the electric field 
    Ey = q*(y-y0)/(r**3)                    # y component of the electric field 
    return Ex,Ey
```

```python
"Case 1: Monopole Electric Field"
#Physical system parameter input
x=np.arange(0,10,1)
y=np.arange(0,10,1)
q1 = -2
x1 = 2
y1 = 2

xx,yy=np.meshgrid(x,y)

# Calculate Field
Ex1,Ey1 = electric_field(x1,y1,xx,yy,q1)



#Ploting the field 
plt.quiver(xx,yy,Ex1,Ey1)
plt.rcParams['figure.figsize']=6,6

plt.xlabel("x")
plt.ylabel('y')
plt.title('Monopole Electric Field')
```

```python
"Case 2: Dipole Electric Field"

q2 = 2
x2 = 8
y2 = 2

#2D array Field Calculation of the second charge
Ex2,Ey2 = electric_field(x2,y2,xx,yy,q2)

#Resulting Field Calculation
Tx12 = Ex1 + Ex2
Ty12 = Ey1 + Ey2

#Ploting the field 
plt.quiver(xx,yy,Tx12,Ty12)
plt.rcParams['figure.figsize']=6,6

plt.xlabel("x")
plt.ylabel('y')
plt.title('Dipole Electric Field')
```

```python
"Case 3: Quadruple Electric Field"

#The 2D array Field Calculation generated by the charges
q3 = -2
x3 = 8
y3 = 8
Ex3,Ey3 = electric_field(x3,y3,xx,yy,q3)

q4 = 2
x4 = 2
y4 = 8
Ex4,Ey4 = electric_field(x4,y4,xx,yy,q4)

Tx1234 = Ex1 + Ex2 + Ex3 + Ex4
Ty1234 = Ey1 + Ey2 + Ey3 + Ey4

#Ploting the field
plt.quiver(xx,yy,Tx1234,Ty1234)
plt.rcParams['figure.figsize']=6,6

plt.xlabel("x")
plt.ylabel('y')
plt.title('Quadruple Electric Field')



```
